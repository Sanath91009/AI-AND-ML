```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("test.csv")
data.shape,display(data)
```

```python
data.dropna(inplace=True)
```

```python
mat=data.iloc[:,:561].values
cov_mat=np.cov(mat,rowvar=False)
```

```python
np.linalg.det(cov_mat)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=100)
pca.fit(mat)
red_mat=pca.transform(mat)
red_data=pd.DataFrame(red_mat)
red_data
```

```python
red_data["subject"]=data["subject"]
red_data["Activity"]=data["Activity"]
```

```python
red_data
```

```python
unq_act=np.unique(np.array(red_data["Activity"]))
act_dict={}
for i in range(len(unq_act)):
    act_dict[unq_act[i]]=i
act_dict
```

```python
labels=[]
for act in red_data["Activity"]:
    labels.append(act_dict[act])
labels=np.array(labels)
```

```python
new_act_dict=dict([(values,key) for key,values in act_dict.items()])
```

```python
new_act_dict
```

```python
from sklearn.metrics import f1_score,accuracy_score
```

```python
from hmmlearn import hmm
def hmm_F1score(N,M,new_act_dict,labels):
    log_likelihood=[]
    for i in range(len(new_act_dict)):
        hmm_act=hmm.GMMHMM(n_components=N,n_mix=M,covariance_type="diag")
        pres_act=red_data[red_data["Activity"]==new_act_dict[i]].iloc[:,0:100].values
        hmm_act.fit(pres_act)
        a=[]
        for j in range(red_data.shape[0]):
            a.append(hmm_act.score(red_data.iloc[j,0:100].values.reshape(1,-1)))
        log_likelihood.append(a)
    pred_label=np.argmax(np.matrix(log_likelihood),axis=0)
    pred_label=np.ravel(pred_label)
    F1=f1_score(labels,pred_label,average="micro")
    acc=accuracy_score(labels,pred_label)
    return F1,acc
```

```python
hmm_F1score(1,1,new_act_dict,labels)
```

```python
import matplotlib.pyplot as plt
F1_state_val=[]
acc_state_val=[]
for i in range(1,7):
    print("HMM trained {} states".format(i))
    f1,acc=hmm_F1score(i,1,new_act_dict,labels)
    F1_state_val.append(f1)
    acc_state_val.append(acc)
fig,axs=plt.subplots(2,1)
axs[0].plot(F1_state_val)    
axs[1].plot(acc_state_val)
```

```python
F1_state_val=[]
acc_state_val=[]
for i in range(1,7):
    print("HMM trained {} states".format(i))
    f1,acc=hmm_F1score(3,i,new_act_dict,labels)
    F1_state_val.append(f1)
    acc_state_val.append(acc)
fig,axs=plt.subplots(2,1)
axs[0].plot(F1_state_val)    
axs[1].plot(acc_state_val)
```
