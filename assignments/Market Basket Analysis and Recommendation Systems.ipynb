```python
import pandas as pd
import numpy as np
```

```python
items=pd.read_csv("GroceryStoreDataSet.csv",header=None)
items["transaction"]=np.arange(items.shape[0])
items
```

```python
items.columns=["items","transaction"]
```

```python
items
```

```python
items_dict={}
total_items=[]
for i in range(items["items"].shape[0]):
    words=items["items"][i].split(",")
    for j in words:
        if j in items_dict :
            items_dict[j].append(i)
        else :
            items_dict[j]=[i]
```

```python
items_dict
```

```python
co_df={}
co_df["transaction"]=np.arange(items["items"].shape[0])
for item in items_dict:
    l=np.zeros(items["items"].shape[0],dtype="int")
    l[items_dict[item]]=1
    co_df[item]=l
```

```python
co_df=pd.DataFrame(co_df).set_index("transaction")
co_df
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_items=apriori(co_df,min_support=0.1,use_colnames=True)
```

```python
freq_items
```

```python
rules=association_rules(freq_items,metric="confidence",min_threshold=0.4)
rules
```
